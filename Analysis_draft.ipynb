**Thoughts**

The light pole point cloud in the K-D tree wastes a lot of space by including the unnecessary base/ground part in half of the tree. This is not reflected in the data flower distribution, posing a risk of compressing some 3-D data into a 2-D representation and carrying along noise data, especially when the higher-dimension data contains significant noise. The data flower representation misleadingly suggests that the light pole data is perfectly divided, whereas half of the entire tree is just noise.

Moreover, expect the phenomenon where the 'tail' or a small outstanding cluster (points clustered in an unusual space) of an object, such as the trunk of a tree or the light of a light pole, would be stored along with the non-outstanding clusters if the tree's depth is shallow.

In this K-D tree, which divides based on the median of the point distribution rather than the space, one may observe a more uniform point distribution in each node of the K-D tree than in an octree. This means there is less chance that the outstanding part of the point cloud will be eliminated with a low density threshold, especially in cases where the point distribution is uneven—some parts of the point cloud may have far fewer points than others, which is common because the scan originates from above, typically resulting in higher point density at the top of an object.

That said, the density threshold affects the K-D tree more slowly than it does an octree, as evidenced by a sudden sharp increase in the number of dead branches among the distributions produced from the inspection of an octree.

However, this doesn't mean the K-D tree is an optimal representation of the entire dataset. Unlike the octree data flowers, which reflect more characteristics of the point cloud objects, the K-D tree data flowers appear uniform across all types of data.
